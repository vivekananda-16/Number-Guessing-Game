```python
import random
import time

MIN_NUMBER = 1
MAX_NUMBER = 200
MAX_GUESSES = 6

def get_player_name():
    print("May I ask you for your name?")
    return input()

def display_intro(name):
    print(f"{name}, we are going to play a game. I am thinking of a number between {MIN_NUMBER} and {MAX_NUMBER}")
    time.sleep(0.5)
    print("Go ahead. Guess!")

def get_valid_guess():
    while True:
        try:
            enter = input("Guess: ")
            guess = int(enter)
            if MIN_NUMBER <= guess <= MAX_NUMBER:
                return guess
            else:
                print("Silly Goose! That number isn't in the range! Please enter a number between 1 and 200")
        except ValueError:
            print(f"I don't think that {enter} is a number. Sorry")

def provide_feedback(guess, number):
    if guess < number:
        print("The guess of the number that you have entered is too low")
    elif guess > number:
        print("The guess of the number that you have entered is too high")

def play_game():
    number = random.randint(MIN_NUMBER, MAX_NUMBER)
    name = get_player_name()
    display_intro(name)
    
    guesses_taken = 0
    while guesses_taken < MAX_GUESSES:
        guess = get_valid_guess()
        guesses_taken += 1

        if guess == number:
            print(f'Good job, {name}! You guessed my number in {guesses_taken} guesses!')
            return

        provide_feedback(guess, number)
        time.sleep(0.5)
        print("Try Again!")

    print(f'Nope. The number I was thinking of was {number}')

if __name__ == "__main__":
    play_again = "yes"
    while play_again.lower() in ["yes", "y"]:
        play_game()
        print("Do you want to play again?")
        play_again = input()
```
